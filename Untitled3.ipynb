```python
# Outliers
Среднее_количество_единиц_в_чеке  = []
for Cреднее_количество_единиц_в_чеке in df['Среднее_количество_единиц_в_чеке']:
    if Cреднее_количество_единиц_в_чеке > (df['Среднее_количество_единиц_в_чеке'].mean() + 3*df['Среднее_количество_единиц_в_чеке'].std()):
        Cреднее_количество_единиц_в_чеке = (dfm['Среднее_количество_единиц_в_чеке'].mean() + 3*df['Среднее_количество_единиц_в_чеке'].std())
    elif Cреднее_количество_единиц_в_чеке < (df['Среднее_количество_единиц_в_чеке'].mean() - 3*df['Среднее_количество_единиц_в_чеке'].std()):
        Cреднее_количество_единиц_в_чеке = (dfm['Среднее_количество_единиц_в_чеке'].mean() - 3*df['Среднее_количество_единиц_в_чеке'].std())
    Cреднее_количество_единиц_в_чеке.append(Cреднее_количество_единиц_в_чеке)
df['Среднее_количество_единиц_в_чеке'] = Cреднее_количество_единиц_в_чеке

Рейтинг_в_гуглі = []
for Рейтинг_в_гуглі in df['Рейтинг_в_гуглі']:
    if Рейтинг_в_гуглі > (df['Рейтинг_в_гуглі'].mean() + 3*df['Рейтинг_в_гуглі'].std()):
        Рейтинг_в_гуглі = (dfm['Рейтинг_в_гуглі'].mean() + 3*df['Рейтинг_в_гуглі'].std())
    elif Рейтинг_в_гуглі < (df['Рейтинг_в_гуглі'].mean() - 3*df['Рейтинг_в_гуглі'].std()):
        Рейтинг_в_гуглі = (dfm['Рейтинг_в_гуглі'].mean() - 3*df['Рейтинг_в_гуглі'].std())
    Рейтинг_в_гуглі.append(Рейтинг_в_гуглі)
df['Рейтинг_в_гуглі'] = Рейтинг_в_гуглі

Автомобілів_в_день = []
for Автомобілів_в_день in df['Автомобілів_в_день']:
    if Автомобілів_в_день > (df['Автомобілів_в_день'].mean() + 3*df['Автомобілів_в_день'].std()):
        Автомобілів_в_день = (dfm['Автомобілів_в_день'].mean() + 3*df['Автомобілів_в_день'].std())
    elif Автомобілів_в_день < (df['Автомобілів_в_день'].mean() - 3*df['Автомобілів_в_день'].std()):
        Автомобілів_в_день = (dfm['Автомобілів_в_день'].mean() - 3*df['Автомобілів_в_день'].std())
    Автомобілів_в_день.append(Автомобілів_в_день)
df['Автомобілів_в_день'] = Автомобілів_в_день

Пішохідний_перехід = []
for Пішохідний_перехід in df['Пішохідний_перехід']:
    if Пішохідний_перехід > (df['Пішохідний_перехід'].mean() + 3*df['Пішохідний_перехід'].std()):
        Пішохідний_перехід = (dfm['Пішохідний_перехід'].mean() + 3*df['Пішохідний_перехід'].std())
    elif Пішохідний_перехід < (df['Пішохідний_перехід'].mean() - 3*df['Пішохідний_перехід'].std()):
        Пішохідний_перехід = (dfm['Пішохідний_перехід'].mean() - 3*df['Пішохідний_перехід'].std())
    Пішохідний_перехід.append(Пішохідний_перехід)
df['Пішохідний_перехід'] = Пішохідний_перехід

```

```python
# Metrics
def metrics(X_train, X_test, y_train, y_test, y_pred, model):
    RMSE = [mean_squared_error(y_train, model.predict(X_train),squared=False), mean_squared_error(y_test, y_pred, squared=False)]
    R2 = [model.score(X_train, y_train), model.score(X_test, y_test)]
    results = {'RMSE': RMSE, 'R2':R2}
    return pd.DataFrame(results, index=['Train', 'Test'])
```
